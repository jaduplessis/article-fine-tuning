```python
from chunks import Chunking

with open('../../../articles/parsed/Serverless webhooks — Designing Lift _ by Frédéric Barthelet _ Serverless Transformation _ Medium.txt') as f:
  doc = f.read()
  print(f"Text length: {len(doc)} characters.")

chunking = Chunking()

chunks = chunking.construct_chunks(doc)
print(f"Number of chunks: {len(chunks)}")
print(f"Average chunk length: {sum([len(c) for c in chunks])/len(chunks)} characters.")
```

```python
for chunk in chunks:
  print(f"Chunk length: {len(chunk)}")
  print(chunk)
  print('\n\n')
```
